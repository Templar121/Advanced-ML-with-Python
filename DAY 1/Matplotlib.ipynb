```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.show()
plt.grid()
plt.figure(figsize = (10, 5))
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
temp = [32, 36, 38, 30, 39, 40, 35]

plt.plot(x,temp)
plt.show()
plt.grid()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
temp = [32, 36, 38, 30, 39, 40, 35]

plt.plot(x,temp)
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Climate")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
temp = [32, 36, 38, 30, 39, 40, 35]

plt.plot(x,temp, marker = "o", color = "red", linestyle = "dashed")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Climate")
plt.grid()
plt.show()
```

```python
company = ['Reliance', 'Adani', 'Tata']
revenue = [110, 120, 90]

xpos = np.arange(len(company))
plt.bar(xpos, revenue)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Comapny Details")
plt.show()
```

```python
company = ['Reliance', 'Adani', 'Tata']
revenue = [110, 120, 90]

xpos = np.arange(len(company))
plt.bar(xpos, revenue, width = 0.2, color = "green", label = "Revenue")
plt.legend()
plt.xticks(xpos, company)
plt.xlabel("Company Name")
plt.ylabel("Revenue")
plt.title("Company Details")
plt.show()
```

```python
x = ['USA', 'China', 'Japan', 'Australia']
G = [40, 40, 20, 18]
S = [44, 27, 12, 19]
B = [42, 24, 13, 16]
b_bronze = np.add(G, S)

plt.bar(x, G, width = 0.4, label = 'Gold', color = "Gold")
plt.bar(x, S, 0.4, bottom = G, label = "Silver", color = "Silver")
plt.bar(x, B, 0.4, bottom = b_bronze, label = "Bronze", color = "brown")
plt.legend()
plt.show()
```

```python
m = ['Apple', 'Oppo', 'Vivo', 'Realme', 'Nokia']
s = [110, 350, 150, 80, 40]

```

```python
plt.pie(s, labels = m)
plt.show()
```

```python
plt.pie(s, labels = m, radius = 0.75, shadow = True)
plt.show()
```

```python
plt.pie(s, labels = m, radius = 0.80, shadow = True,
       autopct = '%1.1f%%')
plt.show()
```

```python
plt.pie(s, labels = m, radius = 0.80, shadow = True,
       autopct = '%1.1f%%', counterclock = True, startangle = 45)
plt.show()
```

```python
plt.pie(s, labels = m, radius = 0.80, shadow = True,
       autopct = '%1.1f%%', counterclock = True, startangle = 45, 
       explode = [0,0,0.3,0,0.4])
plt.legend(loc = "best")
plt.show()
```
