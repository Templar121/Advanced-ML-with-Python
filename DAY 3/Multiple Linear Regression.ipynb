```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('boston.csv')
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
#check the information of the dataset
df.info()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20,8))
ax[0].set_xlabel('Crime')
ax[0].set_ylabel('medv')
ax[0].set_title('Scatter Plotting of Crime with Price')
ax[0].scatter(df['crim'], df['medv'], marker = 'o', color = 'Green')

ax[1].set_xlabel('Crime')
ax[1].set_ylabel('medv')
ax[1].set_title('Histogram Plotting of Crime with Price')
ax[1].hist(df['crim'], color = 'Blue')

plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20,8))
ax[0].set_xlabel('Crime')
ax[0].set_ylabel('medv')
ax[0].set_title('Scatter Plotting of Crime with Price')
ax[0].scatter(df['crim'], df['medv'], marker = 'o', color = 'Green')
ax[0].axis([0, 10, 0, 60])

ax[1].set_xlabel('Crime')
ax[1].set_ylabel('medv')
ax[1].set_title('Histogram Plotting of Crime with Price')
ax[1].hist(df['crim'], color = 'Blue', bins = 100)
ax[1].axis([0, 10, 0, 350])

plt.show()
```

```python
# Redaing the value of the zn
df['zn'].head(5)
```

```python
df['zn'].value_counts()
```

```python
import warnings
warnings.filterwarnings('ignore')

value = ['zero', 'non-zero']
frequencies = [df[df['zn'] == 0]['medv'].mean(),
               df[df['zn'] != 0]['medv'].mean()]
sns.barplot(x = value, y = frequencies)
plt.xlabel('zn')
plt.ylabel('Mean medv')
plt.title('Bar Plot')
plt.show()
```

```python
# indus plotting as per the medv 
plt.figure(figsize = (10, 6))
plt.scatter(df['indus'], df['medv'], color = 'red')
plt.xlabel('industry')
plt.ylabel('medv')
plt.title('indus with medv')
plt.show()
```

```python
# chas analysis with medv
df['chas'].value_counts()
```

```python
value = ['Zero', 'One']
frequencies = [df[df['chas'] == 0].shape[0], 
                    df[df['chas'] == 1].shape[0]]

fig = plt.figure(figsize = (20, 8))
fig.add_subplot(1, 2, 1)
plt.title('Pychart Plotting')
plt.pie(frequencies, labels = value, colors = ['pink', 'green'],
       explode = (0.1, 0), counterclock = False, autopct = '%.2F%%', shadow = True)


fig.add_subplot(1, 2, 2)
plt.title('Bar Plotting')
sns.barplot(x = value, y = frequencies)
plt.show()
```

```python
#nox stand for nitrogen oxide in the environment, Analysis
plt.figure(figsize = (16, 6))
plt.scatter(df['nox'], df['medv'], color = 'gray')
plt.xlabel('Nitrogen oxide')
plt.ylabel('medv')
plt.title('nox v/s medv')
plt.grid(ls = '--')
plt.show()
```
