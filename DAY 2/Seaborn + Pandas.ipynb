### Write a code to create a DataFrame using Numpy and pandas 

```python
np.random.seed(0)
df = pd.DataFrame(dict(time = np.arange(500), value = np.random.randn(500).cumsum()))

df.head()
```

```python
sns.relplot(x = 'time', y = 'value', data = df, kind = 'line')
```

```python
df2 = pd.DataFrame(np.random.randn(500, 2).cumsum(axis = 0), columns = ['time', 'value'])
df2.head()
```

```python
sns.relplot(x= 'time', y = 'value', data = df2, kind = 'line')
```

```python
sns.relplot(x= 'time', y = 'value', data = df2, kind = 'line', sort = False)
```
