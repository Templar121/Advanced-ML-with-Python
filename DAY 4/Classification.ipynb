### The logistic function in linear regression is a type of sigmoid, a class of functions with s0me specific properties. Sigmoid is a mathematical function that takes any reals values and maps to it to probability between 0 to 1 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-5, 5, 1000)
y1 = sigmoid(x)
plt.plot(x, y1, label = 'sigmoid')
plt.legend()
plt.grid()
plt.show()
```

```python
def derivative(x, step):
    return (sigmoid((x + step)) - sigmoid(x))/ step
```

```python

x = np.linspace(-5, 5, 1000)
y1 = sigmoid(x)
y2 = derivative(x, 0.0000000000001)
plt.plot(x, y1, label = 'sigmoid')
plt.plot(x, y2, label = 'derivative')
plt.legend()
plt.grid()
plt.show()
```
