### Data contains one entry for each month for 6 years , so first we need to get rid of day from date



```python
df.dtypes
```

```python
df = pd.read_csv('monthly_price.csv', parse_dates = ['Date'])
```

```python
df.dtypes
```

```python
df['Year'] = df.Date.dt.year
df['Month'] = df.Date.dt.month
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype(str)
df.info()
```

```python
for index, row in df.iterrows():
    df.at[index, 'Date'] = f"{row['Year']}-{row['Month']}"
```

```python
df.head()
```

```python
del df['Year']
```

```python
del df['Month']
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Converting date columns again to date-time format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.Date = df.Date.dt.to_period('m')
```

```python
df.dtypes
```

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=5).mean().plot(color='blue', label='5-period Rolling Mean')
# Add labels and title
plt.title('Price and 5-Period Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

### In timeseries analysis the rolling function is used to calculate rolling or moving-statistic such as movng average(mean) sums, standards deviations, etc. over a specified window of dataframes 

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=5).sum().plot(color='blue', label='5-period Rolling Mean')
# Add labels and title
plt.title('Price and 5-Period Rolling Sum')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=5).max().plot(color='blue', label='5-period Rolling Mean')
# Add labels and title
plt.title('Price and 5-Period Rolling Max')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=5).min().plot(color='blue', label='5-period Rolling Mean')
# Add labels and title
plt.title('Price and 5-Period Rolling Min')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['Price'].plot(color='grey', label='Price')
# Calculate the 5-period rolling mean for the Price column and plot in blue
df['Price'].rolling(window=5).std().plot(color='blue', label='5-period Rolling Mean')
# Add labels and title
plt.title('Price and 5-Period Rolling STD')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
